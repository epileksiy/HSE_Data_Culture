##  Добавляем фичи по комментам

```python
import pickle
# Путь к данным
path = '/Users/fulyankin/Yandex.Disk.localized/Проекты/top_russian_music/comments/youtube/'

# Подгружаем все данные, что у нас есть 
with open(path + 'youtube_sentiments_v2', 'rb') as f:
    youtube_comments = pickle.load(f)
    
len(youtube_comments)
```

```python
youtube_comments[0]
```

```python
# Подгружаем матерные списки 
from tqdm import tqdm_notebook
import requests
import re

url = 'https://github.com/odaykhovskaya/obscene_words_ru/raw/master/obscene_corpus.txt'
response = requests.get(url)

obscene = set([item.lower() for item in response.text.split('\n')])
```

```python
def get_comment_data(cur_name):
    
    comment_features = { }

    try:
        cur_com = [comment for comment in youtube_comments if comment['video_name'] == cur_name]
        
        # комментаторы с уникальными id 
        comment_features['commentators_uniq'] = len(set([com['author'] for com in cur_com]))

        # чилсо символов в комментариях 
        comments_len = [len(com['dirty_text']) for com in cur_com]
        comment_features['comment_mean_len_symbol'] = np.mean(comments_len) # среднее
        comment_features['comment_max_len_symbol'] = np.max(comments_len)   # максимальное

        # число слов в комментариях
        comments_len = [len(com['dirty_text'].split()) for com in cur_com]
        comment_features['comment_mean_len_word'] = np.mean(comments_len)   # среднее
        comment_features['comment_max_len_word'] = np.max(comments_len)     # максимальное

        # негатив в комментариях 
        # (логрегрессия по твиттеру, точность - 0.8, полнота - 0.75 или около того)
        comments_len = [com['twitter_our_model'] for com in cur_com]
        
        # средняя вероятность негатива в комментах под видео 
        comment_features['comment_negative_mean_prob'] = np.mean(comments_len) 
        # средняя доля негатива в комментах под видео при пороге 0.65 
        comment_features['comment_negative_share'] = np.mean(np.array(comments_len) > 0.65) 

        # считаем число матерных слов в коментах 
        obsc = np.array([np.sum([item in obscene for item in com['dirty_text'].lower().strip().split()])
                          for com in cur_com])
        comment_features['words_obscene_cnt'] = np.sum(obsc)         # суммарное число матерных слов
        comment_features['comments_obscene_cnt'] = np.sum(obsc > 0)  # число матерных комментариев 

        # эмодзи и редкие символы 
        emoji_trace = np.array([len(re.sub('[’—\\\xa0\r\n\t0-9ё!"#$%&«»\'()*+,-./:;<=>?@[\\]^_`{|}~а-яa-z ]', '', 
                com['dirty_text'].lower().strip())) for com in cur_com])

        comment_features['comments_emoji_share'] = np.mean(emoji_trace > 0)  # доля комментов с "эмодзи"
        comment_features['comments_emoji_mean'] = np.mean(emoji_trace)       # средняя длина следа из "эмодзи"
        comment_features['comments_emoji_max'] = np.max(emoji_trace)         # максимальная длина следа из "эмодзи"
        
        # название клипа, чтобы потом мёрджить по нему 
        comment_features['title'] = cur_name                
        return comment_features
    
    except:
        print(cur_name)
        return comment_features
```

```python
# Для распараллеливания кода
from joblib import Parallel, delayed

n_jobs = -1 # параллелим на все ядра 
comments_features = Parallel(n_jobs=n_jobs)(delayed(get_comment_data)(
    text) for text in tqdm_notebook(df.title))
```

```python
df_com = pd.DataFrame(comments_features)
print(df_com.shape)
df_com.head()
```

```python
right = df_com.dropna().set_index('title').drop_duplicates()
left = df.set_index('title').drop_duplicates()
result = left.join(right)
result.shape
```

```python
ig, ax = plt.subplots(figsize=(20,12))
sns_heatmap = sns.heatmap(result.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
result.reset_index(inplace=True)
```

```python
result.drop_duplicates(subset = ['video_id'], inplace=True)
result.shape
```

```python
drop_columns = ['Unnamed: 0', 'favoriteCount', 'dimension', 
                'liveBroadcastContent', 'projection']

result.drop(drop_columns, axis=1,inplace=True)
```
