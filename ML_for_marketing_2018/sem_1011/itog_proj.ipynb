# Итоговый проект (как бы страшно это не звучало) 

Во всём курсе на написание кода были только задания на Datacamp. Пришло время исправить это недоразумение и заняться итоговым проектом.  Кекс был призван продемонстрировать то, насколько вы в состоянии работать со смысловой составляющей. Итоговый проект будет направлен на техническую составляющую.

Это задание будет довольно объёмным. В итоговой оценке оно занимает большую долю. Поэтому, дерзайте! :) 

## Что надо делать?

Для решения вам предоставляется несколько задач. Выберете одну из них на свой вкус и цвет. Вам предстоит изучить представленные данные, понять как они устроены и построить модель, которая могла бы помочь решить поставленную задачу. Задачу для решения вы выбираете сами. У вас есть свобода.  

Ваше исследование должно иметь прмерно следущую структуру: 

* Формулирование бизнес-задачи. Что вы хотите сделать? Её перевод на язык машинного обучение, выбор и обоснование метрик качества. 
* Описание данных, их визуальный анализ. 
* Предобработка данных. Чётко опишите какую именно предобработку вам необходимо сделать, а главное поясните зачем. Попробуйте сварить в ходе предобработки новые объясняющие переменные, которые могут помочь в решении задачи.
* Выберете несколько моделей, которые могли бы вам помочь в решении поставленной задачи. Проведите моделирование.
* Постарайтесь достичь максимального качества при моделировании. 

Например, попробуйте перебрать параметры с помощью решётки, попробуйте несколько разных подходов. Сварите новые переменные, кластеризуйте те наблюдения, которыми вы располагаете и используйте кластеры в качестве новой объясняющей переменной (да, это легально). Не забудьте сделать OHE-кодирование... Можете попробовать отобрать только самые важные переменные с помощью Lasso-регрессии и уже на них обучить что-то другое. Можете делать всё, что хотите. Даже гонять нейросетки, если это поможет в решении задачи.

## Оценивание

Проект оценивается по 10-бальной шкале. Все шаги, которые были вами предприняты, должны быть чётко аргументированы. За неаргументированные действия/результаты баллы ставится не будут. Следуйте структуре, описанной выше и всё у вас будет хорошо. Ваша работа будет оцениваться по двум составляющим: написанная вами тетрадка с кодом и выступление.

## Дедлайны

* На первый этап, состоящий в проработке кода, у вас есть $10$ дней, которые стартуют сейчас. Дедлайн $22$ ноября в $23:30$. В это время на почтах семинаристов должны оказаться ваши первые версии тетрадок с кодом. 
* После стартуют $7$ дней, выделенные на доработку результатов и подготовку небольшого выступления. Об этом подробнее вы узнаете через $10$ дней. Они истекают $29$ ноября в $23:30$

## Задачи:

Выберите одну из задач, перечисленных ниже и решите её. 

### [Продажи видеоигр](https://www.kaggle.com/gregorut/videogamesales)

В ваших руках оказались данные по продажам видеоигр. Ваша задача попытаться спрогнозировать объём продаж игры в зависимости от её параметров. Обратите внимание, что данные о продажах фрагментированы по различным регионам. Возможно, это поможет достать из них что-то интересное. 


### [Удержание сотрудников](https://www.kaggle.com/pavansubhasht/ibm-hr-analytics-attrition-dataset/home)

В ваших руках оказались данные по удержанию сотрудников. Ваша задача состоит в предсказании того, уйдёт ли конкретный сотрудник с работы в ближайшее время. Одним из применений данной модели могла бы быть раздача ништяков тем людям, которые в ближайшее время хотят покинуть компанию. Возможно, вы могли бы попробовать понять какие именно ништяки нужно выписать сотрудникам из тестовой выборки, чтобы вероятность их оттока уменьшилась. 


### [Одежда да e-commerce](https://www.kaggle.com/nicapotato/womens-ecommerce-clothing-reviews/home)

В ваших руках оказались данные по электронной коммерции женской одежды. Ключевой составляющей этих данных являются отзывы и рекомендации одежды. По этим данным можно было бы попробовать понять для какой целевой группы женщин подходит какая разновидность товара и выстроить какую-то ульра-крутую стратегию для таргета средствами машинного обучения.


### [Продажи GStore](https://www.kaggle.com/c/ga-customer-revenue-prediction/data)

В ваших руках оказался набор данных по клиентам GoogleMerchandiseStore (также известный как G Store). Вам предстоит спрогнозировать доход магазина на одного клиента. 

В этой задаче вам будут даны грязные логи, из которых вам придётся самостоятельно варить таблички с данными внутренними средствами питона. Если вы решили взять эту задачу, вы очень жёсткий и чёткий чувак.


**Внимание:** Ни в коем случае не пытайтесь заглянуть в будущее! Например, спрогнозировать продажи в Японии сегодня по сегодняшним продажам в Испании. Вы не можете увидеть эти две цифры по очереди. Вы видите их одновременно и, пытаясь построить такой прогноз, заглядываете в будущее.
